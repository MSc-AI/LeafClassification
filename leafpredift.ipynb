```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

from time import time
from sklearn.model_selection import ShuffleSplit, GridSearchCV
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
```

## Import CSV FILES

```python
df_train = pd.read_csv("/Users/isa/Desktop/LeafClassification/train.csv")
df_test = pd.read_csv("/Users/isa/Desktop/LeafClassification/test.csv")
```

```python
df_train.head()
df = df_train.copy()
```

```python
df_test.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedShuffleSplit


encoder=LabelEncoder()
le=encoder.fit(df_train.species)
labels=le.transform(df_train.species)
classes=list(le.classes_)
```

```python
df_train=df_train.drop(['id','species'],axis=1)
test_id=df_test.id
df_test=df_test.drop(['id'],axis=1)
```

```python
df_test
```

```python
uniquelables=np.unique(labels)
```

## Train and Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(df_train,labels,test_size=0.3,shuffle=True,stratify=labels)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pipe_lr=Pipeline([('scl', StandardScaler()), 
                  ('pca', PCA(n_components=10)), 
                  ('clf', LogisticRegression(random_state=1))])
pipe_lr.fit(X_train, y_train)
print(pipe_lr.score(X_test, y_test))
```

## K Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score,KFold
X=df_train
Y=df['species']
logreg=LogisticRegression()
kf=KFold(n_splits=10)
score=cross_val_score(logreg,X,Y,cv=kf)
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

### Most suitable parameter

```python
cv_sets = ShuffleSplit(n_splits=10,test_size=0.30,random_state=42)
classifiers = [SVC(), KNeighborsClassifier()]
params = [{'kernel':('linear','poly','sigmoid','rbf'),'C':[0.01,0.05,0.025,0.07,0.09,1.0], 'gamma':['scale'], 'probability':[True]},
          {'n_neighbors': [3,5,7,9]}]
```

```python
best_estimators = []
for classifier, param in zip(classifiers, params):
    grid = GridSearchCV(classifier,param,cv=cv_sets)
    grid = grid.fit(X_train,y_train)
    best_estimators.append(grid.best_estimator_)
```

```python
best_estimators
```

```python
param_dist = {"max_depth": [3, None],
              "max_features": randint(1, 9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}
tree = DecisionTreeClassifier()

tree_cv = RandomizedSearchCV(tree, param_dist, cv=5)

tree_cv.fit(X_train, y_train)
```
